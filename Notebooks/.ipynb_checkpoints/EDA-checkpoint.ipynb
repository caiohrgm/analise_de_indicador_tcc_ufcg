```python
import pandas as pd
import numpy as np
import warnings

warnings.filterwarnings('ignore')
```

```python
path = r"D:\Documentos_D\UFCG\2021.2e\TCC\Project\Data\indicadores_Siteware.csv"
df = pd.read_csv(path,sep =",")
df.head(2)
```

```python
print("Total de indicadores: ", len(df))
```

```python
filtered_df = df[['Indicador','Dim1_Membro_Mome','Dim2_Membro_Mome','Dim3_Membro_Mome','VLR_Ocorrencia','Tipo_Registro']]
filtered_df = filtered_df[filtered_df['Tipo_Registro'] == "Apurado"].drop(['Tipo_Registro'],axis=1)
```

```python
# Contando NaNs do df filtrado:
dim1_membro_mome_nan_sum = filtered_df['Dim1_Membro_Mome'].isna().sum()
dim2_membro_mome_nan_sum = filtered_df['Dim2_Membro_Mome'].isna().sum()
dim3_membro_mome_nan_sum = filtered_df['Dim3_Membro_Mome'].isna().sum()
vlr_ocorrencia_nan_sum = filtered_df['VLR_Ocorrencia'].isna().sum()

print(dim1_membro_mome_nan_sum, dim2_membro_mome_nan_sum, dim3_membro_mome_nan_sum, vlr_ocorrencia_nan_sum)
```

```python
# Substituindo os valores NaNs pela string 'empty':
filtered_df = filtered_df.fillna("empty")
filtered_df
```

```python
# Removendo do dataframe linhas que possuem as três "Dim_Membro_Mome" vazias:
filtered_df = filtered_df[(filtered_df.Dim1_Membro_Mome != "empty") & (df.Dim2_Membro_Mome != "empty") & 
                          (df.Dim3_Membro_Mome != "empty")]
filtered_df
```

```python
# Removendo do dataframe linhas que possuem o "Valor de Ocorrência" vazio:
filtered_df = filtered_df[(filtered_df.VLR_Ocorrencia != "empty")].reset_index()
```

```python
# Reconfigurando os índices:
new_index = []
for i in range(len(filtered_df)):
    new_index.append(i+1)
filtered_df = filtered_df.set_index(pd.Index(new_index))
filtered_df
```
